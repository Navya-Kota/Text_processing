## Tokenization

```python
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
```

```python
sent=sent_tokenize(text)
sent
```

```python
from nltk import word_tokenize
```

```python
word=word_tokenize(sent[2])
print(sent[2])
print(word)
```

# Removing punctuations

```python
import string
```

```python
from string import punctuation
punctuation
```

```python
pun_words=[]
for w in word:
    if w not in punctuation:
        pun_words.append(w)
pun_words
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop= stopwords.words('english')
print(len(stop))
stop
```

```python
st_words=[]
for w in pun_words:
    if w not in stop:
        st_words.append(w)
st_words
```

# Stemming

```python
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
ps.stem('drinking')
```

```python
for i,w in enumerate(st_words):
    st_words[i]=ps.stem(st_words[i])
st_words
```
